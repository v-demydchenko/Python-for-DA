# Тести для малих вибірок та пропорцій



**Завдання 1**. E-commerce компанія після редизайну сайту підозрює, що **середній час до покупки (time-to-purchase)** користувачів **збільшився**.
Історично середній час від першого заходу на сайт до покупки становив $\mu_0 = 123.7$ хвилин.

Команда аналітиків випадково обрала дані **7 покупців після редизайну**:
`128, 135, 121, 142, 126, 151, 123`

З допомогою наявних даних зʼясуйте, чи збільшився середній час покупки після редизайну.

Для цього виконайте наступні 6 кроків. Правильне виконання кожного з кроків оцінюється в 1 бал.

1. Запишіть нульову та альтернативну гіпотези і визначте тип тесту.
2. Обчисліть вибіркові статистики: $\bar x$, $s$, $n$.
3. Оберіть тип тесту та виконайте його (знайдіть критичне значення тесту, статистику тесту та р-значення) будь-яким пасуючим способом, щоб перевірити гіпотезу на рівні значущості $\alpha = 0.10$.
4. Прийміть рішення, чи відхиляєте ви гіпотезу $H_0$ використовуючи p-value.
5. Напишіть висновок: чи справді редизайн сайту подовжив час до покупки?
6. Чи зміниться ваше рішення при зміні рівня значущості на $\alpha = 0.05$.


1.

*   Нульова гіпотеза (H₀): μ = 123.7 (Редизайн не вплинув на середній час, він залишився колишнім)
*   Альтернативна гіпотеза (Ha): μ > 123.7 (Редизайн збільшив середній час)

Оскільки альтернативна гіпотеза спрямована в один бік (μ > μ₀), це односторонній (one-tailed) t-тест.

2. Обчислення вибіркових статистик

*   n (розмір вибірки): 7
*   x̄ (вибіркове середнє):
    (128 + 135 + 121 + 142 + 126 + 151 + 123) / 7 = 926 / 7 = 132.2857 хв.
*   s (вибіркове стандартне відхилення):
    Спочатку знайдемо суму квадратів відхилень (SS):
    SS = (128-132.2857)² + (135-132.2857)² + (121-132.2857)² + (142-132.2857)² + (126-132.2857)² + (151-132.2857)² + (123-132.2857)²
    SS ≈ (18.37) + (7.35) + (127.51) + (94.51) + (39.51) + (349.51) + (86.51) ≈ 723.27
    s = √(SS / (n-1)) = √(723.27 / 6) = √(120.545) ≈ 10.98 хв.

Відповідь:
*   n = 7
*   x̄ ≈ 132.29 хв.
*   s ≈ 10.98 хв.

3. Виконання t-тесту

Оскільки розмір вибірки малий (n=7 < 30) і стандартне відхилення генеральної сукупності невідоме, використовуємо t-тест.

1.  Ступені свободи (df): df = n - 1 = 7 - 1 = 6
2.  Критичне значення (t_critical):
    Для одностороннього тесту з α = 0.10 та df = 6, критичне значення знаходимо з таблиці t-розподілу:
    t_critical ≈ 1.440
3.  Статистика тесту (t_test):
    Обчислюється за формулою:
    t_test = (x̄ - μ₀) / (s / √n)
    t_test = (132.2857 - 123.7) / (10.98 / √7) = (8.5857) / (10.98 / 2.6458) = 8.5857 / 4.149 ≈ 2.069
4.  p-значення:
    Це ймовірність отримати t-статистику рівну 2.069 або більшу за умови істинності H₀.
    Для t_test = 2.069 та df = 6, p-значення знаходимо за допомогою калькулятора або статистичного ПЗ.
    p-value ≈ 0.042 (або 0.0419)

4. Прийняття рішення за p-value

Правило: якщо p-value ≤ α, то нульову гіпотезу (H₀) відхиляємо.
*   p-value (≈ 0.042) ≤ α (0.10)

Висновок за p-value: Оскільки p-значення менше за рівень значущості 0.10, ми відхиляємо нульову гіпотезу H₀.

5. Висновок

Статистичний аналіз даних (t(6)=2.069, p<0.05) на рівні значущості 10% показує, що існують статистично значущі докази того, що середній час до покупки після редизайну сайту збільшився.

6. Зміна рівня значущості на α = 0.05

*   p-value (≈ 0.042) ≤ α (0.05)

Висновок: Так, p-значення (0.042) все ще менше за новий рівень значущості (0.05). Це означає, що ми також відхилили б нульову гіпотезу на більш строгому рівні значущості 5%. Редизайн дійсно статистично значуще подовжив час до покупки.

**Завдання 2.**

До спеціальної рекламної кампанії **23%** дорослих упізнавали логотип компанії. Після завершення кампанії відділ маркетингу провів опитування: з **1200** випадково відібраних дорослих **311** упізнали логотип.

Перевірте на рівні значущості **3%** ($\alpha=0.03$), чи дають ці дані достатні підстави стверджувати, що **тепер більше ніж 23%** дорослих упізнають логотип компанії. Для розвʼязку використовуйте бібліотеку `statsmodels`.

Зробіть висновок, чи зросла впізнаваність логотипу.

Додатково, обчисліть довірчий інтревал на заданому рівні значущості і проінтерпретуйте текстом - як він додатково пояснює прийняте нами рішення?

```python
import statsmodels.api as sm
from scipy import stats
import numpy as np
```

```python
n = 1200
x = 311
p0 = 0.23
alpha = 0.03

p_hat = x / n
print(f"Вибіркова пропорція (p̂): {p_hat:.4f}")

z_score, p_value = sm.stats.proportions_ztest(count=x, nobs=n, value=p0, alternative='larger')
print(f"z-статистика: {z_score:.4f}")
print(f"p-значення: {p_value:.4f}")

if p_value <= alpha:
    print(f"Рішення: Оскільки p-value ({p_value:.4f}) <= α ({alpha}), ми ВІДХИЛЯЄМО нульову гіпотезу H₀.")
else:
    print(f"Рішення: Оскільки p-value ({p_value:.4f}) > α ({alpha}), ми НЕ МАЄМО ПІДСТАВ відхилити H₀.")

se = np.sqrt(p_hat * (1 - p_hat) / n)           # Стандартна помилка
z_critical = stats.norm.ppf(1 - alpha)          # Критичне значення z
lower_bound = p_hat - z_critical * se           # Нижня межа інтервалу
print(f"97% нижньо-сторонній довірчий інтервал: ({lower_bound:.4f}, 1)")
```
